```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_gexf("../notebooks/test.gexf", relabel=False, version='1.2draft')
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
# Force-Directed
nx.draw_kamada_kawai(G)
```

```python
# Force-Directed pos
nx.spring_layout(G)
```

```python
from nxviz.plots import ArcPlot

a = ArcPlot(G)
a.draw()
plt.show()
```

```python
from nxviz.plots import CircosPlot

c = CircosPlot(G)
c.draw()
plt.show()
```
